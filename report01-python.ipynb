## 制御文：if

`if`文の書式は以下の通りである．

```python:
if 条件:
    処理
```

条件が`True`なら処理が実行される．

条件の内部はインデント（空白4個）で区別される．
インデントはPythonでは必須である．

`else`文を入れて以下のようにもできる．

```python:
if 条件:
    処理1
else:
    処理2
```

もしくは`else`と`if`を連続した`elif`を用いることもできる．

```python:
if 条件1:
    処理1
elif 条件2:
    処理2
elif 条件3:
    処理3
else:
    処理4
```


条件の記述によく使われる比較演算子とその意味を下に記載しておく

- `<=`：以下
- `<`：小さい
- `>=`：以上
- `==`：等しい
- `!=`：等しくない


````python
a = 8

if a <= 10:
    print("a is smaller than or equal to 10")
else:
    print("a is larger than 10")

if a < 10:
    print("a is smaller than 10")
elif a >= 10:
    print("a is larger than or equal to 10")
elif a == 10:
    print("a is equal to 10")
elif a != 10:
    print("a is not equal to 10")
else:
    print("a is a number")
````

論理演算子を使うと，関係演算を組み合わせて条件を記述できる．

- `and`：両方成立するとき
- `or`：どちらかが成立するとき
- `not`：条件の否定




````python
a = 0
b = 10

if a <= 10 and b >= 10:
    print("a <= 10 and b >= 10")

if a < 10 or b > 12:
    print("a < 10 or b > 12")

if not a >= 10:
    print("not a >= 10")
````

## 制御文：for

for文の書式は以下の通りである．

```python:
for 変数 in イテラブル:
    繰り返し処理
```

イテラブルの各要素についてループが実行され，その要素は変数に格納される．

ループの内部はインデント（空白4個）で区別される．
ifと組み合わせる場合には，さらにインデントを追加する．


break文を使うとfor文を途中で抜けることができる．

以下は，長さ5のリストの値を足し合わせるが，もし負の値が入力されたら処理を止める例題である．

````python
sum = 0.0

value_list = [22, 4, 11, 56, 33]
# value_list = [22, 4, -11, 56, 33]

# カウンタiを定義して，一つずつ増やすループ
for i in [0, 1, 2, 3, 4]:
    value = value_list[i]
    if value < 0:  # 入力が負ならループを抜ける    
        print("A negative value is found")
        break

    sum += value

print("Sum: ", sum)
````

上の例ではイテラブルはリスト`[0, 1, 2, 3, 4]`であるが，これは以下のように`range()`関数を使って簡潔に書ける．



````python
sum = 0.0

value_list = [22, 4, 11, 56, 33]
# value_list = [22, 4, -11, 56, 33]

# カウンタiを定義して，一つずつ増やすループ
for i in range(5):
    value = value_list[i]
    if value < 0:
        print("A negative value is found")
        break

    sum += value

print("Sum: ", sum)
````


関数`range()`は
- 引数が1つの場合，引数に与えた整数**未満**の整数リストを生成する．`range(5)`なら`[0, 1, 2, 3, 4]`と同等である．
- 引数が2つの場合，1番目の引数から始まって，2番目の引数に与えた整数**未満**の整数リストを生成する．`range(5, 8)`なら`[5, 6, 7]`と同等である．

なお`range(5)`は`range(0, 5)`と等価である．

````python
print(list(range(5)))
print(list(range(5, 8)))
````

````python
sum = 0.0
print("sum of ", end="")
for i in range(5):
    print(i, ", ", end="")
    sum += i
print("is", sum)


sum = 0.0
print("sum of ", end="")
for i in range(11, 18):
    print(i, ", ", end="")
    sum += i
print("is", sum)
````

なおリスト自体がイテラブルであるため，インデックスの変数`i`を用いずに，直接以下のように書くこともできる．

````python
sum = 0.0

value_list = [22, 4, 11, 56, 33]
# value_list = [22, 4, -11, 56, 33]

# value_listのすべての要素に対してループ
for value in value_list:
    if value < 0:
        print("A negative value is found")
        break

    sum += value

print("Sum: ", sum)
````

## 制御文：while

while文の書式は以下の通りである．

```python:
while 条件:
    繰り返し処理
```

ループの内部はインデント（空白4個）で区別される．


break文を使うとwhile文を途中で抜けることができる．

以下は，長さ5のリストの値を足し合わせるが，もし負の値が入力されたら処理を止める例題である．

````python
sum = 0.0

value_list = [22, 4, 11, 56, 33]
# value_list = [22, 4, -11, 56, 33]

# カウンタiを定義して，一つずつ増やし，リストの長さになるまでループ
i = 0
while i < len(value_list):
    value = value_list[i]
    if value < 0:
        print("A negative value is found")
        break

    sum += value
    i += 1

print("Sum: ", sum)
````

pythonのでの真偽値は`True`と`False`で表される．以下はwhileループを無条件で回し，インデックスが範囲を超えたらbreakする例である．

````python
sum = 0.0

value_list = [22, 4, 11, 56, 33]
# value_list = [22, 4, -11, 56, 33]

# カウンタiを定義して，一つずつ増やし，リストの長さになるまでループ
i = 0
while True:
    value = value_list[i]
    if value < 0:
        print("A negative value is found")
        break

    sum += value

    i += 1
    if i >= len(value_list):
        break


print("Sum: ", sum)
````

## 関数

````python
import numpy as np  # sqrt()を使うために必要なモジュール

def hypotenuse_of_right_triangle(a, b):
    # 直角三角形の斜辺(hypotenuse)を求める関数

    return np.sqrt(a * a + b * b)
````

## 値渡しと参照渡し

pythonの引数の扱いは厳密にはややこしいが（オブジェクトの参照が渡される），以下のように考えて良い．

- 関数の呼び出し時，引数は「値渡し」される（とみなす）


値渡しでは，引数として渡された変数に保持されている値のみがコピーされて関数内に渡されるため，関数内で値を変更しても，呼び出し元の変数には反映されない．

このことは，2つの変数の値を入れ替えるswap関数の作成を考えるとわかりやすい．このプログラムを実行すると，swap前後で変数a, bの値が変わってないことが確認できる．

- [値の変化を可視化するならこのページが便利である](https://pythontutor.com/render.html#code=%23%20%E9%96%A2%E6%95%B0%E5%86%85%E3%81%A7a%E3%81%A8b%E3%81%AE%E5%80%A4%E3%82%92%E5%85%A5%E3%82%8C%E6%9B%BF%E3%81%88%E3%82%8B%0Adef%20swap%28a,%20b%29%3A%0A%20%20%20%20temp%20%3D%20a%0A%20%20%20%20a%20%3D%20b%0A%20%20%20%20b%20%3D%20temp%0A%20%20%20%20print%28%22in%20swap%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%0A%20%20%20%20%0A%0A%23%20swap%E9%96%A2%E6%95%B0%E3%81%AE%E5%91%BC%E3%81%B3%E5%87%BA%E3%81%97%E5%89%8D%E5%BE%8C%E3%81%A7%E5%A4%89%E6%95%B0a,%20b%E3%81%AE%E5%80%A4%E3%82%92%E7%A2%BA%E8%AA%8D%0A%0Aa%20%3D%201%0Ab%20%3D%202%0A%0Aprint%28%22before%20swap%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%0A%0Aswap%28a,%20b%29%0A%0Aprint%28%22after%20swap%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%20%20%23%20%E5%80%A4%E3%81%AF%E5%A4%89%E3%82%8F%E3%81%A3%E3%81%A6%E3%81%84%E3%81%AA%E3%81%84&cumulative=false&curInstr=1&heapPrimitives=nevernest&mode=display&origin=opt-frontend.js&py=3&rawInputLstJSON=%5B%5D&textReferences=false)．（nextとprevのボタンで実行行を移動してみると，右側の変数値が変化する様子がわかる）
  

````python
# 関数内でaとbの値を入れ替える
def swap(a, b):
    temp = a
    a = b
    b = temp
    print("in swap: a = {}, b = {}".format(a, b))
````

````python

# swap関数の呼び出し前後で変数a, bの値を確認

a = 1
b = 2

print("before swap: a = {}, b = {}".format(a, b))

swap(a, b)

print("after swap: a = {}, b = {}".format(a, b))  # 値は変わっていない
````

古典的な言語と異なり，関数は複数の値を（タプルとして）返すことができる．そのため，引数に渡す変数の値を関数内で変更するのではなく，以下のように関数の返り値を複数の変数で（タプルとして）返すことが多い．

````python
def swap(a, b):
    return b, a
````

ただしリストなどの場合には，リスト自体のswapはできないが，中身を変更することはできてしまう．

これは，実際には変数への参照が関数へ渡されるためである．（「参照」とは変数の実体ではなく，実体のありかを指し示すもの）

- [これも変数値の変化を可視化するとわかる](https://pythontutor.com/render.html#code=def%20swap%28a,%20b%29%3A%0A%20%20%20%20return%20b,%20a%0A%20%20%20%20%0Adef%20put_zero%28a,%20b%29%3A%0A%20%20%20%20a%5B0%5D%20%3D%200%0A%20%20%20%20b%5B0%5D%20%3D%200%0A%20%20%20%20print%28%22in%20put_zero%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%0A%20%20%20%20%0Aa%20%3D%20%5B1,%202,%203%5D%0Ab%20%3D%20%5B4,%205,%206%5D%0A%0Aprint%28%22before%20swap%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%0A%0Aswap%28a,%20b%29%20%20%23%20%E4%B8%8A%E3%81%A7%E5%AE%9A%E7%BE%A9%E3%81%97%E3%81%9Fswap%E3%81%AF%E5%BC%95%E6%95%B0%E3%81%8C%E3%83%AA%E3%82%B9%E3%83%88%E3%81%A7%E3%82%82%E5%8F%97%E3%81%91%E4%BB%98%E3%81%91%E3%82%8B%0A%0Aprint%28%22after%20swap%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%20%20%23%20%E5%80%A4%E3%81%AF%E5%A4%89%E3%82%8F%E3%81%A3%E3%81%A6%E3%81%84%E3%81%AA%E3%81%84%0A%0Aput_zero%28a,%20b%29%0A%0Aprint%28%22after%20put_zero%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29&cumulative=false&curInstr=0&heapPrimitives=nevernest&mode=display&origin=opt-frontend.js&py=3&rawInputLstJSON=%5B%5D&textReferences=false)

````python
# 関数内でaとbに0を代入する
def put_zero(a, b):
    a[0] = 0
    b[0] = 0
    print("in put_zero: a = {}, b = {}".format(a, b))
````

````python
# swap関数の呼び出し前後で変数a, bの値を確認

a = [1, 2, 3]
b = [4, 5, 6]

print("before swap: a = {}, b = {}".format(a, b))

swap(a, b)  # 上で定義したswapは引数がリストでも受け付ける

print("after swap: a = {}, b = {}".format(a, b))  # 値は変わっていない

put_zero(a, b)

print("after put_zero: a = {}, b = {}".format(a, b))
````

以下のように`copy()`を用いると，オブジェクトのコピーが生成され，それが関数へ渡されるため，関数内で中身を変更したとしても，呼び出し側の変数には影響を与えないようにできる．

 - [可視化した様子はこちら](https://pythontutor.com/render.html#code=def%20swap%28a,%20b%29%3A%0A%20%20%20%20return%20b,%20a%0A%20%20%20%20%0Adef%20put_zero%28a,%20b%29%3A%0A%20%20%20%20a%5B0%5D%20%3D%200%0A%20%20%20%20b%5B0%5D%20%3D%200%0A%20%20%20%20print%28%22in%20put_zero%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%0A%20%20%20%20%0A%23%20swap%E9%96%A2%E6%95%B0%E3%81%AE%E5%91%BC%E3%81%B3%E5%87%BA%E3%81%97%E5%89%8D%E5%BE%8C%E3%81%A7%E5%A4%89%E6%95%B0a,%20b%E3%81%AE%E5%80%A4%E3%82%92%E7%A2%BA%E8%AA%8D%0A%0Aa%20%3D%20%5B1,%202,%203%5D%0Ab%20%3D%20%5B4,%205,%206%5D%0A%0Aprint%28%22before%20put_zero%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%0A%0Aput_zero%28a.copy%28%29,%20b.copy%28%29%29%0A%0Aprint%28%22after%20put_zero%3A%20a%20%3D%20%7B%7D,%20b%20%3D%20%7B%7D%22.format%28a,%20b%29%29%20%20%23%20%E5%80%A4%E3%81%AF%E5%A4%89%E3%82%8F%E3%81%A3%E3%81%A6%E3%81%84%E3%81%AA%E3%81%84&cumulative=false&curInstr=0&heapPrimitives=nevernest&mode=display&origin=opt-frontend.js&py=3&rawInputLstJSON=%5B%5D&textReferences=false)

````python
# swap関数の呼び出し前後で変数a, bの値を確認

a = [1, 2, 3]
b = [4, 5, 6]

print("before put_zero: a = {}, b = {}".format(a, b))

put_zero(a.copy(), b.copy())

print("after put_zero: a = {}, b = {}".format(a, b))  # 値は変わっていない
````
